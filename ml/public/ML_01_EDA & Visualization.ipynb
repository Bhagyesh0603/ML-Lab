```python
sns.countplot(x='AHD', data=df)
plt.title("Heart Disease Count")
plt.show()
```

```python
df['Age'].hist(bins=10)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
df['Chol'].hist(bins=10)
plt.xlabel("Cholesterol")
plt.ylabel("Count")
plt.title("Cholesterol Distribution")
plt.show()
```

```python
df['RestBP'].hist(bins=10)
plt.xlabel("Resting Blood Pressure")
plt.ylabel("Count")
plt.title("Rest BP Distribution")
plt.show()
```

```python
sns.lineplot(x='AHD', y='Age', data=df)
plt.title("Heart Disease vs Age")
plt.show()
```

```python
sns.catplot(x='AHD', data=df, kind='count')
plt.show()
```

```python
sizes = df['AHD'].value_counts()
sizes

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=['No Heart Disease', 'Heart Disease'],autopct='%1.2f%%',shadow=True)
plt.title("Heart Disease Distribution")
plt.show()
```
